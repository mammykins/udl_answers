# Problem 2.1 Solution

To solve Problem 2.1, we need to calculate the expressions for the slopes (partial derivatives) of the loss function with respect to the parameters ϕ₀ and ϕ₁.

## Step 1: Write down the loss function
The given loss function in equation 2.5 is:
$$
L = \sum_{i=1}^{N} (y_i - (\phi_0 + \phi_1 x_i))^2
$$

## Step 2: Identify the parameters
We need to differentiate the loss function with respect to the parameters ϕ₀ and ϕ₁.

## Step 3: Differentiate the loss function with respect to ϕ₀
First, let's rewrite the loss function using the prediction term \(\hat{y}_i\):
$$
\hat{y}_i = \phi_0 + \phi_1 x_i
$$
Now, the loss function becomes:
$$
L = \sum_{i=1}^{N} (y_i - \hat{y}_i)^2
$$
We need to find the partial derivative of \(L\) with respect to \(\phi_0\):
$$
\frac{\partial L}{\partial \phi_0} = \frac{\partial}{\partial \phi_0} \sum_{i=1}^{N} (y_i - (\phi_0 + \phi_1 x_i))^2
$$
To apply the chain rule, we consider the inner function \( (y_i - (\phi_0 + \phi_1 x_i)) \) and its square:
$$
\frac{\partial L}{\partial \phi_0} = \sum_{i=1}^{N} 2 (y_i - (\phi_0 + \phi_1 x_i)) \cdot \frac{\partial}{\partial \phi_0} (y_i - (\phi_0 + \phi_1 x_i))
$$
Since the derivative of \( (y_i - (\phi_0 + \phi_1 x_i)) \) with respect to \(\phi_0\) is \(-1\), we get:
$$
\frac{\partial L}{\partial \phi_0} = \sum_{i=1}^{N} 2 (y_i - (\phi_0 + \phi_1 x_i)) (-1)
$$
Simplifying this, we have:
$$
\frac{\partial L}{\partial \phi_0} = -2 \sum_{i=1}^{N} (y_i - \phi_0 - \phi_1 x_i)
$$

## Step 4: Differentiate the loss function with respect to ϕ₁
Next, we find the partial derivative of \(L\) with respect to ϕ₁:
$$
\frac{\partial L}{\partial \phi_1} = \frac{\partial}{\partial \phi_1} \sum_{i=1}^{N} (y_i - (\phi_0 + \phi_1 x_i))^2
$$
Again, using the chain rule:
$$
\frac{\partial L}{\partial \phi_1} = \sum_{i=1}^{N} 2 (y_i - (\phi_0 + \phi_1 x_i)) \cdot \frac{\partial}{\partial \phi_1} (y_i - (\phi_0 + \phi_1 x_i))
$$
The derivative of \( (y_i - (\phi_0 + \phi_1 x_i)) \) with respect to \(\phi_1\) is \(-x_i\), so we get:
$$
\frac{\partial L}{\partial \phi_1} = \sum_{i=1}^{N} 2 (y_i - (\phi_0 + \phi_1 x_i)) (-x_i)
$$
Simplifying this, we have:
$$
\frac{\partial L}{\partial \phi_1} = -2 \sum_{i=1}^{N} (y_i - \phi_0 - \phi_1 x_i) x_i
$$

## Final Answer
The expressions for the gradients are:
$$
\frac{\partial L}{\partial \phi_0} = -2 \sum_{i=1}^{N} (y_i - \phi_0 - \phi_1 x_i)
$$
$$
\frac{\partial L}{\partial \phi_1} = -2 \sum_{i=1}^{N} (y_i - \phi_0 - \phi_1 x_i) x_i
$$



### Problem 2.2
Show that we can find the minimum of the loss function in closed-form by setting the expression for the derivatives from problem 2.1 to zero and solving for ϕ₀ and ϕ₁.

#### Answer 2.2

##### Step-by-Step Solution

###### 1. Define the Loss Function
The loss function for linear regression is given by:
$$ L(\phi) = \sum_{i=1}^I (y_i - (\phi_0 + \phi_1 x_i))^2 $$

###### 2. Compute the Derivatives (from Problem 2.1)
The partial derivatives of the loss function with respect to ϕ₀ and ϕ₁ are:
$$ \frac{\partial L}{\partial \phi_0} = -2 \sum_{i=1}^I (y_i - (\phi_0 + \phi_1 x_i)) $$
$$ \frac{\partial L}{\partial \phi_1} = -2 \sum_{i=1}^I x_i (y_i - (\phi_0 + \phi_1 x_i)) $$

###### 3. Set the Derivatives to Zero
To find the minimum of the loss function, we set the derivatives to zero:
$$ \frac{\partial L}{\partial \phi_0} = 0 $$
$$ \frac{\partial L}{\partial \phi_1} = 0 $$

These yield the normal equations:
$$ \sum_{i=1}^I (y_i - (\phi_0 + \phi_1 x_i)) = 0 $$
$$ \sum_{i=1}^I x_i (y_i - (\phi_0 + \phi_1 x_i)) = 0 $$

N.B. A normal equation is a specific form of a system of linear equations used in the context of linear regression. It arises when you set the derivative of the loss function (typically the sum of squared errors) with respect to the model parameters to zero. This process leads to a system of equations that can be solved to find the parameters that minimize the loss function.

In the context of linear regression, the normal equations provide a way to solve for the regression coefficients without using iterative methods like gradient descent.

###### 4. Simplify the Normal Equations
First normal equation:
$$ \sum_{i=1}^I y_i - \sum_{i=1}^I (\phi_0 + \phi_1 x_i) = 0 $$
$$ \sum_{i=1}^I y_i = \phi_0 I + \phi_1 \sum_{i=1}^I x_i $$

Second normal equation:
$$ \sum_{i=1}^I x_i y_i - \sum_{i=1}^I x_i (\phi_0 + \phi_1 x_i) = 0 $$
$$ \sum_{i=1}^I x_i y_i = \phi_0 \sum_{i=1}^I x_i + \phi_1 \sum_{i=1}^I x_i^2 $$

###### 5. Solve the System of Equations
We now have a system of two linear equations:
$$ \sum_{i=1}^I y_i = \phi_0 I + \phi_1 \sum_{i=1}^I x_i $$
$$ \sum_{i=1}^I x_i y_i = \phi_0 \sum_{i=1}^I x_i + \phi_1 \sum_{i=1}^I x_i^2 $$

Let
$$ S_x = \sum_{i=1}^I x_i $$
$$ S_y = \sum_{i=1}^I y_i $$
$$ S_{xx} = \sum_{i=1}^I x_i^2 $$
$$ S_{xy} = \sum_{i=1}^I x_i y_i $$

Then the equations become:
$$ S_y = \phi_0 I + \phi_1 S_x $$
$$ S_{xy} = \phi_0 S_x + \phi_1 S_{xx} $$

Solving for \(\phi_0\) and \(\phi_1\):
$$ \phi_1 = \frac{I S_{xy} - S_x S_y}{I S_{xx} - S_x^2} $$
$$ \phi_0 = \frac{S_y - \phi_1 S_x}{I} $$

Thus, ϕ₀ and ϕ₁ are the closed-form solutions that minimize the least-squares loss function for linear regression.

##### Summary
By setting the derivatives of the loss function to zero and solving the resulting system of equations, we obtain closed-form expressions for the parameters \(\phi_0\) and \(\phi_1\), which can be used to find the best-fitting line for a given set of training data points.

These steps should be verified and solved with actual data points to find the specific values of ϕ₀ and ϕ₁ for a particular dataset. This approach is practical for linear regression problems where a closed-form solution is possible, making it unnecessary to use iterative methods like gradient descent.


### Problem 2.3
Consider reformulating linear regression as a generative model so we have \( x = g[y, ϕ] = ϕ₀ + ϕ₁y \). What is the new loss function? Find an expression for the inverse function \( y = g^{-1}[x, ϕ] \) that we would use to perform inference. Will this model make the same predictions as the discriminative version for a given training dataset {xi, yi}? One way to establish this is to write code that fits a line to three data points using both methods and see if the result is the same.

### Answer 2.3 Reformulating Linear Regression as a Generative Model

#### Step-by-Step Solution

##### 1. Reformulating the Model
We reformulate linear regression such that \( x = g[y, \phi] = \phi_0 + \phi_1 y \).

##### 2. New Loss Function
The new loss function based on the generative model is:
$$ L(\phi) = \sum_{i=1}^I (x_i - (\phi_0 + \phi_1 y_i))^2 $$

##### 3. Inverse Function
To perform inference, we need the inverse function \( y = g^{-1}[x, \phi] \).

Starting with:
$$ x = \phi_0 + \phi_1 y $$

Solving for \( y \):
$$ y = \frac{x - \phi_0}{\phi_1} $$

Thus, the inverse function is:
$$ y = g^{-1}[x, \phi] = \frac{x - \phi_0}{\phi_1} $$

##### 4. Comparing Predictions
To compare the generative and discriminative models, we can fit a line to three data points using both methods and see if the results are the same. Below is the Python code to perform this comparison.


```python
import numpy as np
import matplotlib.pyplot as plt

# Define the data points; data from notebook 2.1
x = np.array([0.03, 0.19, 0.34, 0.46, 0.78, 0.81, 1.08, 1.18, 1.39, 1.60, 1.65, 1.90])
y = np.array([0.67, 0.85, 1.05, 1.0, 1.40, 1.5, 1.3, 1.54, 1.55, 1.68, 1.73, 1.6 ])


# Discriminative model: y = φ₀ + φ₁x
A = np.vstack([x, np.ones(len(x))]).T
phi_discriminative, _, _, _ = np.linalg.lstsq(A, y, rcond=None)
phi_1_discriminative, phi_0_discriminative = phi_discriminative

# Generative model: x = φ₀ + φ₁y
A_gen = np.vstack([y, np.ones(len(y))]).T
phi_generative, _, _, _ = np.linalg.lstsq(A_gen, x, rcond=None)
phi_1_generative, phi_0_generative = phi_generative

# Compare the models
print("Discriminative model parameters (φ₀, φ₁):", phi_0_discriminative, phi_1_discriminative)
print("Generative model parameters (φ₀, φ₁):", phi_0_generative, phi_1_generative)

# Plot the data points and the fitted lines
plt.scatter(x, y, color='blue', label='Data points')
plt.plot(x, phi_0_discriminative + phi_1_discriminative * x, 'r-', label='Discriminative model')
plt.plot((x - phi_0_generative) / phi_1_generative, y, 'g--', label='Generative model')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('Comparison of Discriminative and Generative Models')
plt.show()
```

If the models are equivalent, the parameters and the fitted lines should be the same.
By examining the printed parameters and the plot, we can determine if the generative model makes the same predictions as the discriminative model for the given training dataset.
